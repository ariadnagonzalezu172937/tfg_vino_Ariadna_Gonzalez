# 2020

```python
#Estas lineas de código me permiten crear un file.xlsx por cada hoja que contenga el archivo que quiero leer

df_sheets = pd.ExcelFile('med_2020.xlsx', engine="openpyxl")
hojas_20=[]
for elem in df_sheets.sheet_names:
    df = pd.read_excel('med_2020.xlsx', engine="openpyxl", sheet_name=elem)
    df.to_excel(f"{elem} {'med_2020.xlsx'}.xlsx")
    elem=pd.read_excel(f"{elem} {'med_2020.xlsx'}.xlsx")
    hojas_20.append(elem)
```

```python
#Seleccionamos las columnas que vamos a usar
selected_columns= hojas_20[0].columns[2:7]
selected_columns
```

```python
#limpio cada dataframe, cada una de las hojas del excel que he importado
for i in range(len(hojas_20)):
    hojas_20[i]=hojas_20[i][selected_columns].dropna()
    hojas_20[i]['TEMP'] = hojas_20[i]['TEMP'].astype('float64')
    hojas_20[i]['%RH'] = hojas_20[i]['%RH'].astype('float64')
    
```

```python
#los clasifico por meses y año
meses=['Enero', 'Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre']
df_20=[]
for mes in meses:
    name=mes+'20'
    df_20.append(name)
    name=pd.DataFrame()

df_20[2]=hojas_20[0]
df_20[4]=hojas_20[1]
df_20[5]=hojas_20[2]
df_20[6]=hojas_20[3]
df_20[7]=hojas_20[4]
df_20[8]=hojas_20[5]
df_20[9]=hojas_20[6]
df_20[10]=hojas_20[7]
df_20[11]=hojas_20[8]
```

## Agrupación de datos   

```python
#Creo 3 columnas separadas, para el día, fecha y mes
for i in range(len(meses)):
    if (type(df_20[i])!= str):
        df_20[i]['Date']=[d.date() for d in df_20[i]['TIME']]
        df_20[i]['Time']=[d.time() for d in df_20[i]['TIME']]

for i in range(len(meses)):
    if (type(df_20[i])!= str):
        df_20[i]['Month']= pd.DatetimeIndex(df_20[i]['Date']).month
        
```

```python
df_20[4]
```

```python
#Creo una lista de dataframes para guardar la información de las horas, mientras que trabajo con df_20
info_2020_horas=[]
for i in range(len(meses)):
    name=meses[i]+'20'
    info_2020_horas.append(name)
    if (type(df_20[i])!=str):
        info_2020_horas[i]=df_20[i].drop(['Month'],axis=1)

info_2020_horas[4]
```

```python
#Juntos los datos por días agrupandalos por máximos y minimos en temperatura y humedad
info_2020_dias_max=[]
info_2020_dias_min=[]
for i in range(len(meses)):
    x=meses[i]+'20'
    y=meses[i]+'20'
    if (type(info_2020_horas[i])!= str):
        x =info_2020_horas[i].groupby(['Date']).max()
        y =info_2020_horas[i].groupby(['Date']).min()
        #añado info_mes a info_2020
    info_2020_dias_max.append(x)
    info_2020_dias_min.append(y)
    
for i in range(len(meses)):
    if (type(info_2020_dias_max[i])!= str):
        info_2020_dias_max[i]=info_2020_dias_max[i][['TEMP','%RH']]
        info_2020_dias_max[i]= info_2020_dias_max[i].rename(columns={'TEMP':'TEMP MAX',
                                   '%RH':'HUM MAX'})
        info_2020_dias_min[i]=info_2020_dias_min[i][['TEMP','%RH']]
        info_2020_dias_min[i]=info_2020_dias_min[i].rename(columns={'TEMP':'TEMP MIN',
                                   '%RH':'HUM MIN'})
```

```python
#Agrupo los datos por días realizando medias
info_2020_dias=[]
for i in range(len(meses)):
    info_mes=meses[i]+'20'
    if (type(info_2020_horas[i])!= str):
        info_mes =info_2020_horas[i].groupby(['Date']).mean()
        info_mes= info_mes.rename(columns={'TEMP':'TEMP MED','%RH':'HUM MED'})
        #añado info_mes a info_2020
    info_2020_dias.append(info_mes)
    
#Agrupo la temperatura y la humedad mínima y maxima
for i in range(len(meses)):
    if (type(info_2020_dias[i])!= str):
        info_2020_dias[i]=info_2020_dias[i]
        info_2020_dias[i].insert(0, "TEMP MAX", info_2020_dias_max[i]['TEMP MAX'], True)
        info_2020_dias[i].insert(2, "TEMP MIN", info_2020_dias_min[i]['TEMP MIN'], True)
        info_2020_dias[i].insert(3, "HUM MAX", info_2020_dias_max[i]['HUM MAX'], True)
        info_2020_dias[i].insert(5, "HUM MIN", info_2020_dias_min[i]['HUM MIN'], True)
        
info_2020_dias
```

```python
#Juntos los datos por mes agrupandalos por máximos en temperatura y humedad
info_2020_mes_max=[]
for i in range(len(meses)):
    info_mes=meses[i]+'20'
    if (type(df_20[i])!= str):
        for column in info_2020_dias[i].columns:
            info_mes =df_20[i].groupby(['Month']).max()
        #añado info_mes a info_2020
    info_2020_mes_max.append(info_mes)
    
for i in range(len(meses)):
    if (type(info_2020_mes_max[i])!= str):
        info_2020_mes_max[i]=info_2020_mes_max[i][['TEMP','%RH']]
        info_2020_mes_max[i]= info_2020_mes_max[i].rename(columns={'TEMP':'TEMP MAX',
                                   '%RH':'HUM MAX'})
```

```python
#Juntos los datos por mes agrupandalos por mínimos en temperatura y humedad
info_2020_mes_min=[]
for i in range(len(meses)):
    info_mes=meses[i]+'20'
    if (type(df_20[i])!= str):
        for column in info_2020_dias[i].columns:
            info_mes =df_20[i].groupby(['Month']).min()
        #añado info_mes a info_2020
    info_2020_mes_min.append(info_mes)
    
for i in range(len(meses)):
    if (type(info_2020_mes_min[i])!= str):
        info_2020_mes_min[i]=info_2020_mes_min[i][['TEMP','%RH']]
        info_2020_mes_min[i]= info_2020_mes_min[i].rename(columns={'TEMP':'TEMP MIN',
                                   '%RH':'HUM MIN'})
```

```python
#Agrupo los datos por mes
info_2020_mes=[]
for i in range(len(meses)):
    info_mes=meses[i]+'20'
    if (type(df_20[i])!= str):
        for column in info_2020_dias[i].columns:
            info_mes =df_20[i].groupby(['Month']).mean()
            info_mes= info_mes.rename(columns={'TEMP':'TEMP MED','%RH':'HUM MED'})

        #añado info_mes a info_2020
    info_2020_mes.append(info_mes)
    
#Agrupo la temperatura y la humedad mínima y maxima
for i in range(len(meses)):
    if (type(info_2020_mes[i])!= str):
        info_2020_mes[i]=info_2020_mes[i]
        info_2020_mes[i].insert(0, "TEMP MAX", info_2020_mes_max[i]['TEMP MAX'], True)
        info_2020_mes[i].insert(2, "TEMP MIN", info_2020_mes_min[i]['TEMP MIN'], True)
        info_2020_mes[i].insert(3, "HUM MAX", info_2020_mes_max[i]['HUM MAX'], True)
        info_2020_mes[i].insert(5, "HUM MIN", info_2020_mes_min[i]['HUM MIN'], True)
        
info_2020_mes
```

# Agrupación en una DB

Hasta ahora tenemos 2 grandes bases de datos: los datos aportados por la bodega y los aportados por InfoRiego.

Los datos de la bodega están agrupados por horas, dias y meses entre 2020 y 2023, mientras que los de InfoRiego solo están agrupados por días y meses entre 2018-2021 y solo entre meses para 2022 y 2023.

Los datos de la bodega contieneninformación sobre:
- Temperatura media, máxima y mínima
- Humedad media, máxima y mínima
- WB (bulbo húmedo), contenido de agua en la humedad del aire.
- DP (punto de rocío), temperatura a la cual el aire tiene que ser enfriado antes de que comience la condensación de su humedad

Mientras que los datos de InfoRiego contienen información sobre:
- Temperatura media, máxima y mínima
- Humedad media, máxima y mínima
- Velocidad del viento media y máxima
- Dirección del viento media y máxima
- Radiación
- Precipitación

La idea es crear 3 bases de datos que contengan toda la información para cada año:
- Clasificada por horas db_AÑO_horas
- Clasificada por días db_AÑO_dias
- Clasificada por mes db_AÑO_mes

Dado que los datos aportados por la bodega son más exactos para nuestra predicción usaremos éstos como punto de referencia y añadiremos los otros para rellenar los espacios vacíos sobre todo desde 2018 hasta 2020 y otros tipos de datos que no son recogidos por el sistema de la bodega.
